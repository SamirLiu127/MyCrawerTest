```python
import requests #爬伊利電影下載頁 電影名稱,發文日期,瀏覽人數,網址
from bs4 import BeautifulSoup

print('==============page1==============')
res = requests.get('http://www02.eyny.com/forum-205-1.html')
#print(res.text)
soup = BeautifulSoup(res.text,'lxml')
#print(soup.text)
#print(soup.select('.item'))
i=0
for a in soup.select('tbody'):
    #print(a)
    i += 1
    if i>14:
        print(a.select('.xst')[0].text,
              a.select('.by')[0].select('span')[0].text,
              a.select('em')[2].text,)
        print('http://www02.eyny.com/'+a.select('.xst')[0]['href'])
        
print('==============page2==============')
res = requests.get('http://www02.eyny.com/forum-205-3W0P8JX0.html')
#print(res.text)
soup = BeautifulSoup(res.text,'lxml')
#print(soup.text)
#print(soup.select('.item'))
i=0
for a in soup.select('tbody'):
    #print(a)
    i += 1
    if i>14:
        print(a.select('.xst')[0].text,
              a.select('.by')[0].select('span')[0].text,
              a.select('em')[2].text,)
        print('http://www02.eyny.com/'+a.select('.xst')[0]['href'])
```
